## Reservoirs

```python
reservoirs_file = 'reservoirs_glofas5_03min'
```

```python
reservoirs = xr.open_dataset(PATH / 'old' / f'{reservoirs_file}_20250328.nc')
reservoirs = reservoirs.rename({'Band1': 'reservoirs'})
reservoirs['reservoirs'].attrs['long_name'] = 'reservoirs static map'
reservoirs.close()

# sort latitudes in descending order
reservoirs = reservoirs.reindex(lat=list(reversed(reservoirs.lat)))

reservoirs
```

```python
# export
reservoirs.to_netcdf(PATH / '{0}_{1}.nc'.format(reservoirs_file, datetime.now().strftime('%Y%m%d')))
```
