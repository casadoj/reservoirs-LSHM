### Tables

```python
pars_csv = {}
path_parameters = PATH_EFAS6 / 'reservoirs' / 'tables' / 'parameters'
for file in path_parameters.glob('*.txt'):
    var = file.stem.split('_')[-1]
    serie = pd.read_csv(file, sep='\t', index_col=0, header=None).squeeze()
    serie.name = var
    pars_csv[var] = serie
pars_csv = pd.DataFrame(pars_csv)
pars_csv.index.name = 'GDW_ID'
```
